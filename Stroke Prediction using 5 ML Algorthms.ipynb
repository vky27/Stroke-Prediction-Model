## Now we want to classify the unknown case in smoking class into never smoked,smokes or formerly smoked

```python
tdata = data[data['smoking_status'] != 'Unknown']
tdata
```

```python
## NOW WE WILL APPLY Random forest ON SMOKING STATUS
```

```python
tdata.dropna(inplace=True)
```

```python
tdata.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
tdata.columns
```

```python
X_train = tdata[['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi'
       , 'stroke', 'Male', 'Married', 'Govt_job', 'Private',
       'Self-employed', 'children', 'Urban']]
y_train = tdata['smoking_status']
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train) #Which is close to perfect Score, SO WELL DONE!!
```

```python
udata = data[data['smoking_status'] == 'Unknown']
udata
```

```python
smoking_status = model.predict(udata[['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi', 
        'stroke', 'Male', 'Married', 'Govt_job', 'Private',
       'Self-employed', 'children', 'Urban']])
```

```python
udata['smoking_status'] = smoking_status
```

```python
udata
```

```python
tdata
```

```python
data = pd.concat([tdata,udata],axis=0) 
```

```python
data.head()
```

```python
data.groupby('smoking_status')['stroke'].mean().plot(kind = 'bar') # So Formerly smoked seems more dangerous
```
